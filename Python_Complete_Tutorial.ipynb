
# **3. Python Operations**

There are different kinds of operators in Python.
*   Arithmetic Operators (+, -, *, /, //, %, ***)
*   Logical Operators (and, or, not)
*   Bitwise Operators (&, |, ^, ~, <<, >>)
*   Assignment Operators (=, +=, -=, *=, /=, **=, //=, %=, &=, |=, ^=, ~=, <<=, >>=, :=)
*   Comparison Operators (==, !=, <, >, <=, >=)
*   Identity Operators (is, is not)
*   Membership Operators (in, not in)



```python
# Arithmetic operations
a = 10
b = 5

addition = a + b
subtraction = a - b
multiplication = a * b
division = a / b
floor_division = a // b
modulo = a % b
exponentiation = a ** b

print(f"Addition: {addition}")
print(f"Subtraction: {subtraction}")
print(f"Multiplication: {multiplication}")
print(f"Division: {division}")
print(f"Floor Division: {floor_division}")
print(f"Modulo: {modulo}")
print(f"Exponentiation: {exponentiation}")


# Comparison operations
x = 10
y = 5

equal_to = x == y
not_equal_to = x != y
greater_than = x > y
less_than = x < y
greater_than_or_equal_to = x >= y
less_than_or_equal_to = x <= y

print(f"Equal to: {equal_to}")
print(f"Not equal to: {not_equal_to}")
print(f"Greater than: {greater_than}")
print(f"Less than: {less_than}")
print(f"Greater than or equal to: {greater_than_or_equal_to}")
print(f"Less than or equal to: {less_than_or_equal_to}")


# Logical operations
p = True
q = False

logical_and = p and q
logical_or = p or q
logical_not_p = not p

print(f"Logical AND: {logical_and}")
print(f"Logical OR: {logical_or}")
print(f"Logical NOT p: {logical_not_p}")


# Assignment operations
m = 10
m += 5  # Equivalent to m = m + 5
m -= 2  # Equivalent to m = m - 2
m *= 3  # Equivalent to m = m * 3
m /= 2  # Equivalent to m = m / 2

print(f"After assignment operations: {m}")


# Bitwise operations (example)
n = 10
o = 5
bitwise_and = n & o
print(f"Bitwise AND: {bitwise_and}")
bitwise_or = n | o
print(f'Bitwise OR: {bitwise_or}')
bitwise_xor = n ^ o
print(f'Bitwise XOR: {bitwise_xor}')
bitwise_not_n = ~n
print(f'Bitwise NOT n: {bitwise_not_n}')
```


# **4. Python Decision Statements if-elif-else**

If-else conditional statements are the decision statements where we build a condition whose choices are boolean. Either the condition is being fulfilled or not. So, either True (1) or False (0).
<br><br>
We can use conditionals in Python in four ways:
*   if statement
*   if-else statement
*   if-elif-else statement
*   Nested if-statements

# decision-statement-flowchart.jpeg

```python
# Only if statement
age = 20
if age >= 18:
    print("You are an adult.")

# if-else statement
age = 15
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")

# if-elif-else statement
score = 85
if score >= 90:
    print("A Grade")
elif score >= 80:
    print("B Grade")
elif score >= 70:
    print("C Grade")
else:
    print("Below C")

# Nested if-else statement
age = 25
has_license = True

if age >= 16:
    if has_license:
        print("You can drive.")
    else:
        print("You are old enough but do not have a license.")
else:
    print("You are too young to drive.")
```


# **5. Python Loops while-for**

Loops are a feature of programming that helps us to execute some lines of code several times without us writing those lines again and again.
<br><br>
These are also helpful if we want to iterate over the contents of a sequence, i.e., a List or a String.
<br><br>
In Python, we use two types of loops:
1.   ***While Loop:***  When we want specific lines of code to repeat if a particular condition is met. We can also use 'break' (to come out of the loop at a certain point) and 'continue' (to skip a particular iteration) for making more personalised usage of the loop.
2.   ***For Loop:***  When we need to priorly define exactly how many times we want to iterate a set of lines, or we need to iterate over a sequence (List/String), or we need to track a value inside a sequence, we particularly use For loop as it makes it very easy to handle such conditions.




loops-python.jpeg

```python
# Example 1: For loop iterating over a list
colors = ["red", "green", "blue"]
for color in colors:
    print(color)

# Example 2: For loop with range() function
for i in range(5):  # Iterates from 0 to 4
    print(i)

# Example 3: For loop with range() and step
for i in range(1, 10, 2):  # Iterates from 1 to 9 with a step of 2
    print(i)

# Example 4: While loop
count = 0
while count < 3:
    print(count)
    count += 1

# Example 5: Nested loops (for loop inside another for loop)
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")

# Example 6: Looping through a dictionary
person = {"name": "Bob", "age": 25, "city": "New York"}
for key, value in person.items():
    print(f"{key}: {value}")
```


# **6. Python Functions**

Functions are used to increase the reusability of a chunk of code without causing redundancy. We define a functionality/feature in some lines of code and keep them inside a function. Whenever we need to solve a problem involving that same functionality defined in our function, we just make a function call for that function and don't have to write all that code again.

# Python-Function-Syntax.webp

```python
def greet(name):
  """Greets the person passed in as a parameter."""
  print(f"Hello, {name}!")

greet("World") # call the function


def add_numbers(x, y):
  """Adds two numbers and returns the sum."""
  return x + y

sum_result = add_numbers(5, 3)
print(f"The sum is: {sum_result}")


def is_even(number):
  """Checks if a number is even."""
  return number % 2 == 0

number = 10
if is_even(number):
  print(f"{number} is even")
else:
  print(f"{number} is odd")
```

The pre-defined function contained in libraries of a programming language that come along when we install the packages of a programming language on our local system, are called build-in Functions of that programming language. The built-in functions contain several functionalities, and we get an advantage by calling them for the type of functionality we need, which is contained in that build-in function.

built-in-functions-python.png

```python
# Example usage of built-in functions

# print() - Prints any value paseed to it
print("Hello, World!") # Output: 'Hello, World!'


# input() - Reads a line from input
name = input("Enter your name: ") # Output: 'Enter your name: '
print(f"Hello, {name}!")  # Output: 'Hello, <name>!'


# len() - Returns the length of an object
print(len("Hello"))  # Output: 5


# type() - Returns the type of an object
print(type(5))  # Output: <class 'int'>


# abs() - Returns the absolute value of a number
print(abs(-5))  # Output: 5


# all() - Returns True if all elements of an iterable are true
print(all([True, True, True]))  # Output: True
print(all([True, False, True])) # Output: False


# any() - Returns True if any element of an iterable is true
print(any([False, False, True])) # Output: True
print(any([False, False, False])) # Output: False


# bin() - Converts an integer to its binary representation
print(bin(10))  # Output: 0b1010


# bool() - Converts a value to a Boolean
print(bool(0))  # Output: False
print(bool(1))  # Output: True


# chr() - Returns the character represented by a Unicode
print(chr(65))  # Output: A


# dir() - Returns a list of valid attributes for an object
print(dir([])) # Output: list of attributes


# divmod() - Returns the quotient and remainder of a division
print(divmod(10, 3))  # Output: (3, 1)


# enumerate() - Adds a counter to an iterable
for index, value in enumerate(['a', 'b', 'c']):
  print(index, value) # Output: index and value for each item


# eval() - Evaluates a string as a Python expression
print(eval('2 + 3'))  # Output: 5


# exec() - Executes a string as a Python statement
exec('print("Hello, World!")') # Output: 'Hello, World!'


# float() - Converts a value to a floating-point number
print(float('3.14'))  # Output: 3.14


# int() - Converts a value to an integer
print(int('10'))  # Output: 10


# isinstance() - Checks if an object is an instance of a class
print(isinstance(5, int))  # Output: True


# list() - Creates a list from an iterable
print(list("hello"))  # Output: ['h', 'e', 'l', 'l', 'o']


# map() - Applies a function to each item of an iterable
def square(x):
  return x ** 2
print(list(map(square, [1, 2, 3])))  # Output: [1, 4, 9]


# max() - Returns the largest item in an iterable
print(max([1, 5, 2]))  # Output: 5


# min() - Returns the smallest item in an iterable
print(min([1, 5, 2]))  # Output: 1


# next() - Retrieves the next item from an iterator
my_iterator = iter([1, 2, 3])
print(next(my_iterator))  # Output: 1


# oct() - Converts an integer to its octal representation
print(oct(10))  # Output: 0o12


# ord() - Returns the Unicode of a character
print(ord('A'))  # Output: 65


# pow() - Raises a number to a power
print(pow(2, 3))  # Output: 8


# range() - Generates a sequence of numbers
for i in range(3):
  print(i)  # Output: 0, 1, 2


# reversed() - Returns a reversed iterator
for i in reversed([1, 2, 3]):
  print(i)  # Output: 3, 2, 1


# round() - Rounds a number to a specified number of digits
print(round(3.14159, 2)) # Output: 3.14


# set() - Creates a set object
print(set([1, 2, 2, 3]))  # Output: {1, 2, 3}


# sorted() - Returns a sorted list from an iterable
print(sorted([3, 1, 4, 1, 5, 9, 2, 6]))  # Output: [1, 1, 2, 3, 4, 5, 6, 9]


# str() - Converts a value to a string
print(str(123))  # Output: "123"


# sum() - Sums the items of an iterable
print(sum([1, 2, 3])) # Output: 6


# tuple() - Creates a tuple from an iterable
print(tuple([1, 2, 3])) # Output: (1, 2, 3)
```


# **8. Error/Exception Handling in Python**

Errors are a part of our coding life. We can't get rid of them. Actually, for some people, errors bring beauty and interest in a program.  
<br><br>
There are 3 major and general types of errors in programming:


1.   ***Syntax Error:***  (An error in the spelling of keywords, incorrect placements of keywords, identifiers, or special characters involved in writing a code in a specific language as defined in the rules of that language. It is highlighted before the code compilation.)
2.   ***Runtime Error:***  (An error which occurs while interpreting the code when the program execution is halted because some requirement is not met at a certain point.)
3.   ***Logical Error:***  (An error that doesn't cause any trouble in program execution, but the given problem is not solved by it, i.e., we don't get the required output (solution).)

<br>

Now, we have several types of Runtime Errors in Python, and we don't want our program to be halted abruptly while facilitating the User. For this, we have to handle different types of errors and generate user-friendly error messages that the user can understand, so that they may enter the appropriate input values or guess why their required solution can't be facilitated.
<br><br>
The different types of Runtime Errors are:


1.   ZeroDivisionError
2.   NameError
3.   AttributeError
4.   ValueError
5.   IndexError
6.   TypeError
7.   KeyError
8.   FileNotFoundError
9.   ImportError

```python
# ZeroDivisionError
try:
  result = 10 / 0   # Any number can't be divided by 0.
except ZeroDivisionError:
  print("Division by zero error")

# TypeError
try:
  result = "hello" + 10   # An integer can't be concatenated with a string
except TypeError:
  print("Type error")

# NameError
try:
  print(undefined_variable)   # We are trying to print a variable which we never created.
except NameError:
  print("Name error")

# IndexError
try:
    my_list = [1, 2, 3]
    print(my_list[5])   # We are trying to approach an index of a list which doesn't exist.
except IndexError:
    print("Index out of bounds")

# ValueError
try:
  int("abc")    # Trying to convert a value to integer, which is not a number.
except ValueError:
  print("Value error")

# KeyError
try:
    my_dict = {"a": 1, "b": 2}
    print(my_dict["c"])     # Trying to approach a key, value pair which doesn't exist.
except KeyError:
    print("Key not found in the dictionary")

# FileNotFoundError
try:
    with open("nonexistent_file.txt", "r") as f:   # Trying to open a file which doesn't exist.
        contents = f.read()
except FileNotFoundError:
    print("File not found")

# ImportError
try:
  import non_existent_module     # Trying to import a module which doesn't exist.
except ImportError:
  print("Import error")

# AttributeError
try:
  my_list = [1,2,3]
  my_list.non_existent_method()    # Trying to call a method with an object, which doesn't belong to the class of the object.
except AttributeError:
  print("Attribute error")
```


# **9. File Handling in Python**

In proficient programming practices, we gather input from files, perform functionality, process it, and then save the output in a file.
<br><br>
In most companies, the hiring tests involve this practice too. So, File Handling is a very essential part of programming. Each programming language has a different syntax and different ways of handling files. Like in Python, we have built-in function "open()" and a keyword "with" to handle files for reading, writing, appending, and closing a file.

file-handling.webp

```python
def basic_file_handling_examples():
    try:
        # 0. File Creating and Writing (write mode - also creates a new file)
        with open("my_file.txt", "w") as file:
            file.write("This is a new file.\n")
            print("File created and written (write mode).")

        # 1. File Opening (read mode)
        with open("my_file.txt", "r") as file:  # Use 'with' for automatic closing
            contents = file.read()
            print("File contents (read):\n", contents)

        # 2. File Writing (write mode - overwrites existing file)
        with open("my_file.txt", "w") as file:
            file.write("This is some new text.\n")
            print("File written (write mode).")

        # 3. Appending to File (append mode)
        with open("my_file.txt", "a") as file:
            file.write("This text is appended.\n")
            print("File updated (append mode).")

        # 4. Reading lines
        with open("my_file.txt", "r") as file:
            lines = file.readlines()
            print("File contents (lines):")
            for line in lines:
                print(line.strip())  # Use strip to remove extra whitespace

        # 5. Iterating over lines
        with open("my_file.txt", "r") as file:
            print("File contents (iterating over lines):")
            for line in file:
                print(line.strip())

    except FileNotFoundError:
        print("File not found.")

    except Exception as e:
        print(f"An error occurred: {e}")


basic_file_handling_examples()
```

# **1. Python Lists**

Python lists are like ordered containers that can hold a collection of items. These items can be of different data types, such as numbers, text (strings), or even other lists! You can easily add, remove, and access items within a list, making them incredibly versatile for organizing and manipulating data in your Python programs.
<br><br>
Some predefined methods for "list" class's objects are:


*   append()
*   insert()
*   remove()
*   pop()
*   index()
*   count()
*   sort()
*   reverse()




python-list.jpeg

python-list-slicing-indexing.jpeg

List-Slicing.jpeg

```python
# Python List Operations

my_list = [1, 2, 3, 4, 5]

# Append an element
my_list.append(6)
print("List after appending 6:", my_list)   # Output: List after appending 6: [1, 2, 3, 4, 5, 6]

# Insert an element at a specific index
my_list.insert(2, 7)
print("List after inserting 7 at index 2:", my_list)  # Output: List after inserting 7 at index 2: [1, 2, 7, 3, 4, 5, 6]

# Remove an element by value
my_list.remove(3)
print("List after removing 3:", my_list)  # Output: List after removing 3: [1, 2, 7, 4, 5, 6]

# Remove an element by index
my_list.pop(0) # removes element at index 0
print("List after popping element at index 0:", my_list)   # Output: List after popping element at index 0: [2, 7, 4, 5, 6]

# Find the index of an element
index = my_list.index(5)
print("Index of 5:", index)  # Output: Index of 5: 3


# Count the occurrences of an element
count = my_list.count(4)
print("Number of times 4 appears:", count)  # Output: Number of times 4 appears: 1


# Sort the list
my_list.sort()
print("Sorted list:", my_list)  # Output: Sorted list: [2, 4, 5, 6, 7]

# Reverse the list
my_list.reverse()
print("Reversed list:", my_list)  # Output: Reversed list: [7, 6, 5, 4, 2]


# List slicing
sliced_list = my_list[1:4]  # elements from index 1 up to (but not including) index 4
print("Sliced list (index 1 to 3):", sliced_list)  # Output: Sliced list (index 1 to 3): [4, 5, 6]


# List concatenation
list1 = [1, 2, 3]
list2 = [4, 5, 6]
concatenated_list = list1 + list2
print("Concatenated list:", concatenated_list)  # Output: Concatenated list: [1, 2, 3, 4, 5, 6]


# List comprehension
squares = [x**2 for x in range(1, 6)]
print("List of squares:", squares)   # Output: List of squares: [1, 4, 9, 16, 25]

# Check if element exists in list
if 5 in my_list:
  print("5 is present in the list")
```

```python
# Set Operations

# Create a set
my_set = {1, 2, 3, 4, 5}
print("Original set:", my_set)   # Output: Original set: {1, 2, 3, 4, 5}

# Add an element
my_set.add(6)
print("Set after adding 6:", my_set)  # Output: Set after adding 6: {1, 2, 3, 4, 5, 6}

# Remove an element (raises KeyError if not present)
my_set.remove(3)  # Or use discard() to avoid KeyError if element not present
print("Set after removing 3:", my_set)  # Output: Set after removing 3: {1, 2, 4, 5, 6}

# Discard an element (does not raise an error if not present)
my_set.discard(7)  # No error even if 7 is not present
print("Set after discarding 7:", my_set)  # Output: Set after discarding 7: {1, 2, 4, 5, 6}


# Remove and return an arbitrary element
removed_element = my_set.pop()
print("Removed element:", removed_element)  # Output: Removed element: 1
print("Set after popping an element:", my_set)  # Output: Set after popping an element: {2, 4, 5, 6}


# Clear the set
my_set.clear()
print("Set after clearing:", my_set)  # Output: Set after clearing: set()


# Set operations (union, intersection, difference, symmetric difference)
set1 = {1, 2, 3}
set2 = {3, 4, 5}

# Union
union_set = set1 | set2
print("Union of set1 and set2:", union_set)   # Output: Union of set1 and set2: {1, 2, 3, 4, 5}


# Intersection
intersection_set = set1 & set2
print("Intersection of set1 and set2:", intersection_set)   # Output: Intersection of set1 and set2: {3}


# Difference (elements in set1 but not in set2)
difference_set = set1 - set2
print("Difference of set1 and set2:", difference_set)   # Output: Difference of set1 and set2: {1, 2}


# Symmetric difference (elements in either set1 or set2, but not both)
symmetric_difference_set = set1 ^ set2
print("Symmetric difference of set1 and set2:", symmetric_difference_set)  # Output: Symmetric difference of set1 and set2: {1, 2, 4, 5}


# Check if a set is a subset or superset of another set
set3 = {2, 3}
is_subset = set3.issubset(set1)
print("Is set3 a subset of set1?", is_subset)   # Output: Is set3 a subset of set1? True

is_superset = set1.issuperset(set3)
print("Is set1 a superset of set3?", is_superset)   # Output: Is set1 a superset of set3? True

# Check if sets are disjoint (no common elements)
are_disjoint = set1.isdisjoint(set2) # Note, set1 and set2 share the element 3.
print(f'Are set1 and set2 disjoint? {are_disjoint}')  # Output: Are set1 and set2 disjoint? False


# Create an empty set
empty_set = set()

# Copy a set
set_copy = set1.copy()
print("Copied set:", set_copy)  # Output: Copied set: {1, 2, 3}


# Check if element exists in set
if 5 in my_set:
  print("5 is present in the set")  # Output: 5 is present in the set
else:
  print("5 is not present in the set")  # Output: 5 is not present in the set
```


# **3. Python Tuple**
Python tuples are like ordered, immutable collections of items. This means that once a tuple is created, its elements cannot be changed or modified. Tuples are often used to represent fixed collections of data, such as coordinates (x, y), or to ensure that the order of elements remains consistent.
<br><br>
The only 2 methods for the "tuple" class's objects are:

*   count()
*   index()

python-tuple.jpeg

diff-list-tuple.jpeg

```python
# Tuple functionalities

# Creating a tuple
my_tuple = (1, 2, 3, "apple", 4.5)
print("My tuple:", my_tuple)  # Output: My tuple: (1, 2, 3, 'apple', 4.5)

# Accessing elements
print("First element:", my_tuple[0])  # Output: First element: 1
print("Last element:", my_tuple[-1])  # Output: Last element: 4.5

# Slicing
print("Elements from index 1 to 3:", my_tuple[1:4])  # Output: Elements from index 1 to 3: (2, 3, 'apple')

# Length of the tuple
print("Length of the tuple:", len(my_tuple))  # Output: Length of the tuple: 5

# Concatenation
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
concatenated_tuple = tuple1 + tuple2
print("Concatenated tuple:", concatenated_tuple)  # Output: Concatenated tuple: (1, 2, 3, 4, 5, 6)

# Repetition
repeated_tuple = tuple1 * 3
print("Repeated tuple:", repeated_tuple)  # Output: Repeated tuple: (1, 2, 3, 1, 2, 3, 1, 2, 3)

# Membership check
print("Is 'apple' in the tuple?", "apple" in my_tuple)  # Output: Is 'apple' in the tuple? True

# Iteration
for item in my_tuple:
    print(item)

# Tuple methods
print("Number of times 'apple' appears:", my_tuple.count("apple"))  # Output: Number of times 'apple' appears: 1
print("Index of 'apple':", my_tuple.index("apple"))  # Output: Index of 'apple': 3
```


# **4. Python Dictionary**
Python dictionaries are like collections of key-value pairs. Each key in a dictionary is unique and is associated with a corresponding value. Dictionaries are incredibly useful for storing and retrieving data efficiently, as you can quickly access a value by its key. Think of them like a real-world dictionary where you look up a word (key) to find its definition (value).
<br><br>
Some of the methods for "dict" class's objects are:


*   get()
*   pop()
*   items()
*   keys()
*   values()
*   clear()
*   copy()
*   fromkeys()
*   update()
*   setdefault()









dictionaries-in-python.jpeg

# python dict.webp

```python
# Dictionary functionalities

# Creating a dictionary
my_dict = {"name": "Alice", "age": 30, "city": "New York"}
print("Original dictionary:", my_dict)  # Output: Original dictionary: {'name': 'Alice', 'age': 30, 'city': 'New York'}


# Accessing values
print("Name:", my_dict["name"])  # Accessing using key
# Output: Name: Alice
print("Age:", my_dict.get("age"))  # Accessing using get() method (returns None if key not found)
# Output: Age: 30


# Adding key-value pairs
my_dict["occupation"] = "Engineer"
print("Dictionary after adding occupation:", my_dict)  # Output: Dictionary after adding occupation: {'name': 'Alice', 'age': 30, 'city': 'New York', 'occupation': 'Engineer'}


# Updating values
my_dict["age"] = 31
print("Dictionary after updating age:", my_dict)  # Output: Dictionary after updating age: {'name': 'Alice', 'age': 31, 'city': 'New York', 'occupation': 'Engineer'}


# Removing key-value pairs
del my_dict["city"]  # Removing using del keyword
print("Dictionary after deleting city:", my_dict)  # Output: Dictionary after deleting city: {'name': 'Alice', 'age': 31, 'occupation': 'Engineer'}

occupation = my_dict.pop("occupation") # Removing and returning the value associated with the key
print("Removed occupation:", occupation)  # Output: Removed occupation: Engineer
print("Dictionary after popping occupation:", my_dict)  # Output: Dictionary after popping occupation: {'name': 'Alice', 'age': 31}

# Checking if a key exists
if "name" in my_dict:
    print("The 'name' key exists.")

# Iterating through a dictionary
for key in my_dict:
    print(key, ":", my_dict[key])  # Output: name : Alice
                                   #          age : 31

for key, value in my_dict.items():  # Output: name : Alice
    print(key, ":", value)          #          age : 31

# Get all keys
keys = my_dict.keys()
print("Keys:", keys)  # Output: Keys: dict_keys(['name', 'age'])

# Get all values
values = my_dict.values()
print("Values:", values)   # Output: Values: dict_values('Alice', 31)


# Clear the dictionary
my_dict.clear()
print("Dictionary after clearing:", my_dict) # Output: {}

# Create a copy of the dictionary
original_dict = {"a": 1, "b": 2}
copied_dict = original_dict.copy()
copied_dict["c"] = 3
print(f'Original dict: {original_dict}')   # Output: Original dict:
print(f'Copied dict: {copied_dict}')

# fromkeys() method
keys = ['a', 'b', 'c']
value = 0
new_dict = dict.fromkeys(keys, value)
print(f"New dictionary from keys: {new_dict}")

# update() method
d1 = {"a":1, "b":2}
d2 = {"c":3, "d":4}
d1.update(d2)
print(f'Updated dictionary: {d1}')

#setdefault
my_dict = {"a": 1, "b": 2}
print(my_dict.setdefault("a", 3))  # Output: 1 (value already present)
print(my_dict) # Output: {'a': 1, 'b': 2}

print(my_dict.setdefault("c", 3))  # Output: 3 (new value is set)
print(my_dict) # Output: {'a': 1, 'b': 2, 'c': 3}
```

# **1. Python Class & Object (instance)**


class-template.png

```python
# Define a class named Dog
class Dog:
    # Constructor (initializer)
    def __init__(self, name, breed):
        self.name = name  # Instance variable
        self.breed = breed # Instance variable

    # Method to make the dog bark
    def bark(self):
        print("Woof!")

    # Method to describe the dog
    def describe(self):
        print(f"My name is {self.name} and I am a {self.breed}.")

# Create instances (objects) of the Dog class
dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Lucy", "Labrador")


# Accessing attributes and calling methods
print(dog1.name)  # Output: Buddy
dog1.bark()      # Output: Woof!
dog1.describe()   # Output: My name is Buddy and I am a Golden Retriever.

print(dog2.name)  # Output: Lucy
dog2.bark()      # Output: Woof!
dog2.describe()   # Output: My name is Lucy and I am a Labrador.
```


# **2. Inheritance**

single-inheritance-in-python-1024x615.webp

# inheritance-example.jpeg

types of inheritance.jpeg

```python
# Define a parent class (base class)
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("Generic animal sound")

# Define a child class (derived class) that inherits from Animal
class Dog(Animal):
    def speak(self):
        print("Woof!")

# Define another child class that inherits from Animal
class Cat(Animal):
    def speak(self):
        print("Meow!")

# Create instances of the classes
animal = Animal("Generic Animal")
dog = Dog("Buddy")
cat = Cat("Whiskers")

# Call the speak() method on each instance
animal.speak()  # Output: Generic animal sound
dog.speak()     # Output: Woof!
cat.speak()     # Output: Meow!
```


# **3. Polymorphism**

polymorphism-python.jpeg

# python-polymorphism.png

```python
class Person:
    def __init__(self, name):
        self.name = name

    def role(self):
        print(f"{self.name}'s role is undefined.")

class Boy(Person):
    def role(self):
        print(f"{self.name} is a boy.")

class Friend(Person):
    def role(self):
        print(f"{self.name} is a friend.")

class Student(Person):
    def role(self):
        print(f"{self.name} is a student.")

class Child(Person):
    def __init__(self, name, parents):
        super().__init__(name)
        self.parents = parents

    def role(self):
        print(f"{self.name} is a child of {' and '.join(self.parents)}.")


# Example usage
person = Person("Generic Person")
person.role()  # Output: Generic Person's role is undefined.

boy = Boy("Max")
boy.role()  # Output: Max is a boy.

friend = Friend("Alice")
friend.role() # Output: Alice is a friend

student = Student("Bob")
student.role() # Output: Bob is a student

child = Child("Charlie", ["John", "Jane"])
child.role() # Output: Charlie is a child of John and Jane.


# Polymorphism in action:
people = [person, boy, friend, student, child]
for person in people:
  person.role()
```


# **4. Abstraction**

python-abstraction.png

```python
# Abstraction in Python

from abc import ABC, abstractmethod

class Shape(ABC): # Abstract base class
  @abstractmethod
  def area(self):
    pass

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  # overriding the method defined in the base class
  def area(self):
    return 3.14 * self.radius * self.radius

class Rectangle(Shape):
  def __init__(self, width, height):
    self.width = width
    self.height = height

  # overriding the method defined in the base class
  def area(self):
    return self.width * self.height

# Create objects
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Access area method (demonstrates abstraction)
print("Circle area:", circle.area())
print("Rectangle area:", rectangle.area())
```

encaps-python.jpeg

python-encapsulation-example.png

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self._account_number = account_number  # Protected attribute
        self._balance = balance  # Protected attribute

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited ${amount}. New balance: ${self._balance}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self._balance}")
        else:
            print("Insufficient funds or invalid withdrawal amount.")

    def get_balance(self):
        return self._balance  # Accessing protected attribute through a method


# Example usage
account = BankAccount("1234567890", 1000)

# Accessing attributes directly (not recommended, but possible)
# print(account._account_number)  # Output: 1234567890
# print(account._balance)  # Output: 1000

# Accessing through methods
account.deposit(500)  # Output: Deposited $500. New balance: $1500
account.withdraw(200)  # Output: Withdrew $200. New balance: $1300
print(account.get_balance())  # Output: 1300

account.withdraw(2000) # Output: Insufficient funds or invalid withdrawal amount
```
